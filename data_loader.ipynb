```python
## point cloud
Xrec_dict = {}
X0_dict = {}
Xrec_name = {'Xrec_0','Xrec_1','Xrec_2','Xrec_3','Xrec_4','Xrec_5'}
X0_name = ('X0_0','X0_1','X0_2','X0_3','X0_4','X0_5',)
for i in range(6):
    file = 'point_cloud_'+str(i+1)+'.mat'
    data = loadmat(file)
    Xrec_name = 'Xrec_'+str(i)
    X0_name = 'X0_'+str(i)
    Xrec_dict[Xrec_name] = data['Xrec']
    X0_dict[X0_name] = data['X0']


size = 10
plt.figure(figsize=(8,5))
plt.subplot(2,3,1)
plt.scatter(Xrec_dict['Xrec_0'][:,0],Xrec_dict['Xrec_0'][:,1],s=size,c='r',marker='o',label='generated')
plt.scatter(X0_dict['X0_0'][:,0],X0_dict['X0_0'][:,1],s=size,c='b',marker='+',label='training data')
plt.xlim([-2,2])
plt.ylim([-2,2])
# plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
# plt.xticks([])
# plt.yticks([])
plt.legend()
plt.subplot(2,3,2)
plt.scatter(Xrec_dict['Xrec_1'][:,0],Xrec_dict['Xrec_1'][:,1],s=size,c='r',marker='o',label='generated')
plt.scatter(X0_dict['X0_1'][:,0],X0_dict['X0_1'][:,1],s=size,c='b',marker='+',label='training data')
plt.xlim([-2,2])
plt.ylim([-0.5,1.5])
# plt.xlabel('$X_1$')
# plt.ylabel('$X_2$')
plt.legend()
plt.subplot(2,3,3)
plt.scatter(Xrec_dict['Xrec_2'][:,0],Xrec_dict['Xrec_2'][:,1],s=size,c='r',marker='o',label='generated')
plt.scatter(X0_dict['X0_2'][:,0],X0_dict['X0_2'][:,1],s=size,c='b',marker='+',label='training data')
plt.xlim([-2,2])
plt.ylim([-0.5,1.5])
# plt.xlabel('$X_1$')
# plt.ylabel('$X_2$')
plt.legend()
plt.subplot(2,3,4)
plt.scatter(Xrec_dict['Xrec_3'][:,0],Xrec_dict['Xrec_3'][:,1],s=size,c='r',marker='o',label='generated')
plt.scatter(X0_dict['X0_3'][:,0],X0_dict['X0_3'][:,1],s=size,c='b',marker='+',label='training data')
plt.xlim([-2,2])
plt.ylim([-2,1])
plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
plt.legend()
plt.subplot(2,3,5)
plt.scatter(Xrec_dict['Xrec_4'][:,0],Xrec_dict['Xrec_4'][:,1],s=size,c='r',marker='o',label='generated')
plt.scatter(X0_dict['X0_4'][:,0],X0_dict['X0_4'][:,1],s=size,c='b',marker='+',label='training data')
plt.xlim([0,2])
plt.ylim([0,1])
plt.xlabel('$X_1$')
# plt.ylabel('$X_2$')
plt.legend()
plt.subplot(2,3,6)
plt.scatter(Xrec_dict['Xrec_5'][:,0],Xrec_dict['Xrec_5'][:,1],s=size,c='r',marker='o',label='generated')
plt.scatter(X0_dict['X0_5'][:,0],X0_dict['X0_5'][:,1],s=size,c='b',marker='+',label='training data')
plt.xlim([0,2])
plt.ylim([0,1])
plt.xlabel('$X_1$')
# plt.ylabel('$X_2$')
plt.legend()
fig = plt.gcf()
fig.savefig('./point_cloud.png')
```
