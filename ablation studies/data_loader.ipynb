```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# comparison of different weighting functions
x_cem_weight = np.load("curve_cem_lambda.npy")
x1_data_cem_weight = np.load("curve_cem_x1_data_lambda.npy")
x2_data_cem_weight = np.load("curve_cem_x2_data_lambda.npy")

x_ddpm_weight = np.load("curve_cem_lambda2.npy")
x1_data_ddpm_weight = np.load("curve_cem_x1_data_lambda2.npy")
x2_data_ddpm_weight = np.load("curve_cem_x2_data_lambda2.npy")

x_constant_weight = np.load("curve_cem_constant.npy")
x1_data_constant_weight = np.load("curve_cem_x1_data_constant.npy")
x2_data_constant_weight = np.load("curve_cem_x2_data_constant.npy")
```

```python
## scatter plot
plt.figure(figsize=(11,2.5))
plt.subplot(1,4,1)
plt.scatter(x_constant_weight[:,0], x_constant_weight[:,1])
# plt.xlim([-15,15])
# plt.ylim([-15,15])
plt.xlabel('$X_1$')
plt.ylabel('$X_2$')
# plt.xticks([])
# plt.yticks([])
# plt.title('New score')
plt.subplot(1,4,2)
plt.scatter(x_ddpm_weight[:,0], x_ddpm_weight[:,1])
# plt.xlim([-15,15])
# plt.ylim([-15,15])
plt.xlabel('$X_1$')
# plt.xticks([])
# plt.yticks([])
# plt.title('Score')
plt.subplot(1,4,3)
plt.scatter(x_cem_weight[:,0], x_cem_weight[:,1])
# plt.xlim([-15,15])
# plt.ylim([-15,15])
plt.xlabel('$X_1$')
# plt.xticks([])
# plt.yticks([])
# plt.title('DDPM')
plt.subplot(1,4,4)
plt.scatter(x1_data_cem_weight, x2_data_cem_weight)
# plt.xlim([-15,15])
# plt.ylim([-15,15])
plt.xlabel('$X_1$')
# plt.xticks([])
# plt.yticks([])
# plt.title('DDPM')
# plt.show()
fig = plt.gcf()
fig.savefig('./comparison_weighting_func.png',dpi=300, bbox_inches = 'tight')
```
